```python
import matplotlib.pyplot as plt

total_atoms = 20000
atoms = {
    '219At': total_atoms,
    '215Bi': 0,
    '219Rn': 0,
    '211Pb': 0,
    '211Bi': 0,
    '207Tl': 0,
    '207Pb': 0,
    '2017Pb': 0
}
decay_constants = {
    '219At': 0.97 / 56,
    '219Rn': 0.03 / 56,
    '215Bi_to_211Pb': 1 / 7.6,
    '219Rn_to_211Pb': 1 / 4,
    '211Pb_to_211Bi': 1 / 31.1,
    '211Bi_to_207Tl': 0.997 / 2.14,
    '211Bi_to_2017Pb': 0.003 / 2.14,
    '207Tl_to_207Pb': 1 / 4.77
}

simulation_time = 400
time_points = []
isotope_counts = {isotope: [] for isotope in atoms.keys()}

time = 0
while time <= simulation_time:
    time_points.append(time)
    for isotope in atoms.keys():
        decay_rate = atoms[isotope] * decay_constants.get(isotope, 0)
        atoms[isotope] -= decay_rate
        if isotope == '219At':
            atoms['215Bi'] += decay_rate
        elif isotope == '219Rn':
            atoms['211Pb'] += decay_rate
        elif isotope == '215Bi':
            atoms['211Pb'] += decay_rate
        elif isotope == '211Pb':
            atoms['211Bi'] += decay_rate
        elif isotope == '211Bi':
            atoms['207Tl'] += decay_rate
            atoms['2017Pb'] += decay_rate
        elif isotope == '207Tl':
            atoms['207Pb'] += decay_rate

        isotope_counts[isotope].append(atoms[isotope])

    time += 1

plt.figure(figsize=(10, 6))
for isotope, counts in isotope_counts.items():
    plt.plot(time_points, counts, label=isotope)

plt.xlabel('Time (minutes)')
plt.ylabel('Number of Atoms')
plt.legend()
plt.title('Decay of Isotopes')
plt.grid(True)
plt.show()
```
